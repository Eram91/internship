```python
#1. python program to display all the header tags from wikipedia.org

from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import requests
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(html, "html.parser")
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])

df= pd.DataFrame({'list of header tags': titles})
df
```

```python
#2. Python program to display list of respected former presidents of India

from bs4 import BeautifulSoup
import requests
import pandas as pd
POI = requests.get('https://presidentofindia.nic.in/former-presidents')
POI
bs = BeautifulSoup(POI.content)
presidents = []
for i in bs.find_all('div', class_="desc-sec"):
    presidents.append(i.text)
presidents

df= pd.DataFrame({'list of former presidents of India':presidents})
df
```

```python
#6. Write a python program to scrape the details of most downloaded articles from AI in last 90days.

from bs4 import BeautifulSoup
import requests
import pandas as pd

articles = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
Bsoup = BeautifulSoup(articles.content, 'html.parser')

title = []
for i in Bsoup.find_all('h2', class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    title.append(i.text)
title

authors = []
for i in Bsoup.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)
authors

published = []
for i in Bsoup.find_all('span', class_="sc-1thf9ly-2 dvggWt"):
    published.append(i.text)
published

dataF=pd.DataFrame({'Paper title':title, 'Authors':authors, 
                   'Published Date':published})
dataF
```

```python
#7. A python program to scrape mentioned details from dineout.co.in

from bs4 import BeautifulSoup
import requests
import pandas as pd

dineout = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
beautS = BeautifulSoup(dineout.content)

restaurant = []
for i in beautS.find_all('div', class_="restnt-info cursor"):
    restaurant.append(i.text)
restaurant

cuisine = []
for i in beautS.find_all('div', class_="detail-info"):
    cuisine.append(i.text)
cuisine

location = []
for i in beautS.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)
location

ratings = []
for i in beautS.find_all('div', class_="restnt-rating rating-4"):
    ratings.append(i.text)
ratings

images = []
for i in beautS.find_all("img", class_="no-img"):
    images.append(i['data-src'])
images

data=pd.DataFrame({'Restaurant name':restaurant, 'Cuisine':cuisine, 
                   'location':location, 'Ratings':ratings, 'Image URL':images})

data
```
